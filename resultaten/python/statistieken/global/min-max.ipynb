```python
import pandas as pd
merged=pd.read_csv("/Users/lennertbontinck/Documents/github/bachelorproef-compressie/resultaten/csv/merged.csv")
```

```python
max_sharpness = merged.groupby(['filename', 'extension'], sort=False)['sharpness'].max().to_frame()
max_sharpness = max_sharpness.rename(columns = {'sharpness':'max_sharpness'})

min_sharpness = merged.groupby(['filename', 'extension'], sort=False)['sharpness'].min().to_frame()
min_sharpness = min_sharpness.rename(columns = {'sharpness':'min_sharpness'})

max_color_contrast =  merged.groupby(['filename', 'extension'], sort=False)['color_contrast'].max().to_frame()
max_color_contrast = max_color_contrast.rename(columns = {'color_contrast':'max_color_contrast'})

min_color_contrast =  merged.groupby(['filename', 'extension'], sort=False)['color_contrast'].min().to_frame()
min_color_contrast = min_color_contrast.rename(columns = {'color_contrast':'min_color_contrast'})

max_general =  merged.groupby(['filename', 'extension'], sort=False)['general'].max().to_frame()
max_general = max_general.rename(columns = {'general':'max_general'})

min_general =  merged.groupby(['filename', 'extension'], sort=False)['general'].min().to_frame()
min_general = min_general.rename(columns = {'general':'min_general'})
```

```python
merged_min_max=max_sharpness.merge(min_sharpness,on=['filename', 'extension'])
merged_min_max=merged_min_max.merge(max_color_contrast,on=['filename', 'extension'])
merged_min_max=merged_min_max.merge(min_color_contrast,on=['filename', 'extension'])
merged_min_max=merged_min_max.merge(max_general,on=['filename', 'extension'])
merged_min_max=merged_min_max.merge(min_general,on=['filename', 'extension'])
merged_min_max
```
